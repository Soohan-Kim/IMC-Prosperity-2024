```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools

pd.options.display.float_format = '{:.6f}'.format
```

```python
currencies = ['1', '2', '3']
all_combinations = itertools.product(currencies, repeat=4)
filtered_combinations = [
    combo for combo in all_combinations if not any(combo[i] == combo[i+1] for i in range(len(combo)-1))
]
filtered_combinations
```

```python
exchange_rates = {
    'shells': {'1': 1.41, '2': 0.61, '3': 2.08, 'shells': 1},
    '1': {'1': 1, '2': 0.48, '3': 1.52, 'shells': 0.71},
    '2': {'1': 2.05, '2': 1, '3': 3.26, 'shells': 1.56},
    '3': {'1': 0.64, '2': 0.3, '3': 1, 'shells': 0.46}}
start_currency = 'shells'

currencies = ['1', '2', '3']

profitable_trades = []
for comb in filtered_combinations:
    amount = 1
    current_currency = start_currency

    for currency in comb:
        # Multiply the amount by the rate to the next currency
        print(comb, currency, amount)
        amount *= exchange_rates[current_currency][currency]
        current_currency = currency
        
    # Finally, convert back to shells
    amount *= exchange_rates[current_currency]['shells']
    print(comb, currency, amount)
    if amount > 1:
        profitable_trades.append((comb, amount))

# sort profitable trades by amount
profitable_trades.sort(key=lambda x: x[1], reverse=True)
profitable_trades
```
