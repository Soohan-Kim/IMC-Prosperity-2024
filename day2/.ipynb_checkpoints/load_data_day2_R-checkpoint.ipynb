```r
library(psych)
library(MASS)
```

```r
price <- read.csv("price.csv")
price <- price[order(price$DAY),]
head(price)
```

```r
describe(price)
```

```r
# function for lag creation 
create_lag_vars <- function(data, var_names, lag){
  for (var in var_names) {
    for (i in 1:lag) {
      data[[paste0(var, "_lag_", i)]] <- c(rep(NA, i), head(data[[var]], -i))}}
  return(data)}
```

```r
vars_to_lag <- c("SUNLIGHT", "HUMIDITY")
lag <- 100
```

```r
df_lagged <- create_lag_vars(price, vars_to_lag, lag)
outcome <- price$ORCHIDS
df_lagged
```

```r
df <- subset(df_lagged, select = -c(DAY, timestamp))
```

```r
clean_df <- na.omit(df)
n <- nrow(clean_df)

# Forward stepwise regression
null <- lm(ORCHIDS ~ 1, data=clean_df)
full <- lm(ORCHIDS ~ . , data=clean_df)
fwd <- step(null, scope=formula(full), direction="forward", k=log(n))
summary(fwd)
```

```r
clean_df$production_change <- ifelse(clean_df$SUNLIGHT < 7, -4 * ((7 - clean_df$SUNLIGHT) / 10), 0)
clean_df$production_change <- clean_df$production_change +
  ifelse(clean_df$HUMIDITY < 60 | clean_df$HUMIDITY > 80, -2 * abs((clean_df$HUMIDITY - 70) / 5), 0)

clean_df$shipping_cost <- clean_df$TRANSPORT_FEES +clean_df$EXPORT_TARIFF + clean_df$IMPORT_TARIFF
clean_df$sunlight_less_than_7hrs <- ifelse(clean_df$SUNLIGHT < 4200, 1, 0)
clean_df$humidity_difference_outside_range <- ifelse(clean_df$HUMIDITY < 60 | clean_df$HUMIDITY > 80, abs(clean_df$HUMIDITY - 70), 0)

head(clean_df)
```

```r
# Forward stepwise regression
null <- lm(ORCHIDS ~ 1, data=clean_df)
full <- lm(ORCHIDS ~ . , data=clean_df)
fwd <- step(null, scope=formula(full), direction="forward", k=log(n))
summary(fwd)
```

```r
regression_results <- lm(ORCHIDS ~ sunlight_less_than_7hrs + shipping_cost + shipping_cost
                         + humidity_difference_outside_range, data = clean_df)
summary(regression_results)
```

```r
regression_results <- lm(ORCHIDS ~ SUNLIGHT + HUMIDITY + IMPORT_TARIFF + EXPORT_TARIFF + TRANSPORT_FEES 
                         + sunlight_less_than_7hrs + humidity_difference_outside_range, data = clean_df)
summary_regression <- summary(regression_results)
coefficients <- summary_regression$coefficients
coefficients
```

```r
train_size <- round(0.8 * nrow(clean_df))
train_data <- clean_df[1:train_size, ]
test_data <- clean_df[(train_size + 1):nrow(clean_df), ]

# Fit the regression model using training data
regression_model <- lm(ORCHIDS ~ sunlight_less_than_7hrs + shipping_cost + shipping_cost
                         + humidity_difference_outside_range, data = train_data)

# Predict on testing data
predictions <- predict(regression_model, newdata = test_data)

# Evaluate the model (you can use different metrics depending on your goal)
# For example, you can calculate RMSE (Root Mean Squared Error)
rmse <- sqrt(mean((test_data$ORCHIDS - predictions)^2))
print(paste("RMSE:", rmse))

mae <- mean(abs(test_data$ORCHIDS - predictions))
print(paste("MAE:", mae))
```
