```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.6f}'.format
```

```python
price_0 = pd.read_csv('prices_round_1_day_0.csv', delimiter=';')
grouped_price_0 = price_0.groupby('product')
grouped_price_0.apply(lambda x: x.describe())
```

```python
price_m1 = pd.read_csv('prices_round_1_day_-1.csv', delimiter=';')
grouped_price_m1 = price_m1.groupby('product')
grouped_price_m1.apply(lambda x: x.describe())
```

```python
price_m2 = pd.read_csv('prices_round_1_day_-2.csv', delimiter=';')
grouped_price_m2 = price_m2.groupby('product')
grouped_price_m2.apply(lambda x: x.describe())
```

```python
trade_0 = pd.read_csv('trades_round_1_day_0_nn.csv', delimiter=';')
grouped_trade_0 = trade_0.groupby('symbol')
grouped_trade_0.apply(lambda x: x.describe())
```

```python
for product, group in grouped_trade_0:
    plt.figure()  # Create a new figure for each product
    plt.plot(group.index, group['price'])
    plt.ylim(group['price'].min() - group['price'].std()*2, group['price'].max() + group['price'].std()*2)  
    plt.xlabel('Index')
    plt.ylabel('Price')
    plt.title(f'Price for {product}')
    plt.grid(True)
    
# Show plot
plt.show()
```
