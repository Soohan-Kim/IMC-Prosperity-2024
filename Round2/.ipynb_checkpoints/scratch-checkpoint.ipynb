```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
day_neg1 = pd.read_csv('./data/round-2-island-data-bottle/prices_round_2_day_-1.csv')
day_0 = pd.read_csv('./data/round-2-island-data-bottle/prices_round_2_day_0.csv')
day_1 = pd.read_csv('./data/round-2-island-data-bottle/prices_round_2_day_1.csv')
```

```python
day_neg1
```

```python
def extract_df(df):
    cols = df.columns[0].split(';')
    def allocate_cols(inp):
        return inp.split(';')

    df[df.columns[0]] = df[df.columns[0]].apply(allocate_cols)
    df = df[df.columns[0]].tolist()
    df = pd.DataFrame(df, columns=cols).astype(float)
    
    return df
```

```python
df1 = extract_df(day_neg1)
df1
```

```python
df2 = extract_df(day_0)
df2
```

```python
df3 = extract_df(day_1)
df3
```
