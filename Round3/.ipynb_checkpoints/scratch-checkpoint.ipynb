```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

root = './round-3-island-data-bottle'
```

```python
def allocate_cols(inp):
    return inp.split(';')
```

```python
def get_data(filename):
    df = pd.read_csv(root + '/' + filename)
    cols = df.columns[0].split(';')
    df[df.columns[0]] = df[df.columns[0]].apply(allocate_cols)
    df = df[df.columns[0]].tolist()
    df = pd.DataFrame(df, columns=cols)
    df_g = df[df['product'] == 'GIFT_BASKET'].reset_index(drop=True).drop(columns=['product']).replace('', '0').astype(float)
    df_r = df[df['product'] == 'ROSES'].reset_index(drop=True).drop(columns=['product']).replace('', '0').astype(float)
    df_s = df[df['product'] == 'STRAWBERRIES'].reset_index(drop=True).drop(columns=['product']).replace('', '0').astype(float)
    df_c = df[df['product'] == 'CHOCOLATE'].reset_index(drop=True).drop(columns=['product']).replace('', '0').astype(float)
    
    return {'GIFT_BASKET': df_g, 'ROSES': df_r, 'STRAWBERRIES': df_s, 'CHOCOLATE': df_c}
```

```python
fs = ['prices_round_3_day_' + str(i) + '.csv' for i in range(3)]
fs
```

```python
d = {}

for f in fs:
    d[int(f.split('_')[-1][0])] = get_data(f)
```

```python
d[0]['GIFT_BASKET']
```

```python
y = d[0]['GIFT_BASKET']['mid_price'].pct_change().dropna()
x1 = d[0]['ROSES']['mid_price'].pct_change().dropna()
x2 = d[0]['STRAWBERRIES']['mid_price'].pct_change().dropna()
x3 = d[0]['CHOCOLATE']['mid_price'].pct_change().dropna()
X = pd.concat([x1, x2, x3], axis=1)

#X = sm.add_constant(X)
model1 = sm.OLS(y, X)
results1 = model1.fit()

results1.summary()
```

```python
residuals = results1.resid

# Plot histogram of residuals
plt.hist(residuals, bins=20, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(y, label='true-basket')
plt.plot(results1.predict(X), label='ols-replication')
plt.legend()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(y - results1.predict(X))
```

```python
plt.figure(figsize=(20, 15))
plt.plot(X@np.array([1/11, 6/11, 4/11]), label='portfolio')
plt.plot(y, label='true-basket')
plt.legend()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(y - X@np.array([1/11, 6/11, 4/11]))
```

```python
y = d[0]['GIFT_BASKET']['mid_price']
x1 = d[0]['ROSES']['mid_price']
x2 = d[0]['STRAWBERRIES']['mid_price']
x3 = d[0]['CHOCOLATE']['mid_price']
X = pd.concat([x1, x2, x3], axis=1)

#X = sm.add_constant(X)
model11 = sm.OLS(y, X)
results11 = model11.fit()

results11.summary()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(y - results11.predict(X), label='true')
```

```python
residuals = results11.resid

# Plot histogram of residuals
plt.hist(residuals, bins=20, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sa = y - X@np.array([1, 6, 4])
plt.plot(sa, label='diff-series')
plt.plot([sa.mean()]*len(sa), label='mean')
plt.plot([sa.mean() + sa.std()]*len(sa))
plt.plot([sa.mean() - sa.std()]*len(sa))
plt.plot([sa.mean() + 2*sa.std()]*len(sa))
plt.plot([sa.mean() - 2*sa.std()]*len(sa))
plt.legend()
```

```python
# For all days
y = pd.concat([d[i]['GIFT_BASKET']['mid_price'] for i in range(3)], axis=0).reset_index(drop=True)
x1 = pd.concat([d[i]['ROSES']['mid_price'] for i in range(3)], axis=0).reset_index(drop=True)
x2 = pd.concat([d[i]['STRAWBERRIES']['mid_price'] for i in range(3)], axis=0).reset_index(drop=True)
x3 = pd.concat([d[i]['CHOCOLATE']['mid_price'] for i in range(3)], axis=0).reset_index(drop=True)

X = pd.concat([x1, x2, x3], axis=1)

plt.clf()

plt.figure(figsize=(20, 15))
sa = y - X@np.array([1, 6, 4])
plt.plot(sa, label='diff-series')
plt.plot([sa.mean()]*len(sa), label='mean')
plt.plot([sa.mean() + sa.std()]*len(sa))
plt.plot([sa.mean() - sa.std()]*len(sa))
plt.plot([sa.mean() + 2*sa.std()]*len(sa))
plt.plot([sa.mean() - 2*sa.std()]*len(sa))
plt.legend()
```
